# Visualizations

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
test = pd.read_csv('test.csv')
test['Date'] = pd.to_datetime(test['Date'])
test.set_index('Date', inplace=True)

first_test_day = test.index[0]
```

## SARIMA Predictions

```python
SARIMA_1 = pd.read_csv('predictions/predictions_SARIMA/SARIMA_1.csv')
SARIMA_5 = pd.read_csv('predictions/predictions_SARIMA/SARIMA_5.csv')
SARIMA_21 = pd.read_csv('predictions/predictions_SARIMA/SARIMA_21.csv')

SARIMA_1['Date'] = pd.to_datetime(SARIMA_1['Unnamed: 0'])
SARIMA_1.set_index('Date', inplace=True)
SARIMA_5['Date'] = pd.to_datetime(SARIMA_5['Unnamed: 0'])
SARIMA_5.set_index('Date', inplace=True)
SARIMA_21['Date'] = pd.to_datetime(SARIMA_21['Unnamed: 0'])
SARIMA_21.set_index('Date', inplace=True)

SARIMA_1_ex = pd.read_csv('predictions/predictions_SARIMA/SARIMA_1_ex.csv')
SARIMA_5_ex = pd.read_csv('predictions/predictions_SARIMA/SARIMA_5_ex.csv')
SARIMA_21_ex = pd.read_csv('predictions/predictions_SARIMA/SARIMA_21_ex.csv')

SARIMA_1_ex['Date'] = pd.to_datetime(SARIMA_1_ex['Unnamed: 0'])
SARIMA_1_ex.set_index('Date', inplace=True)
SARIMA_5_ex['Date'] = pd.to_datetime(SARIMA_5_ex['Unnamed: 0'])
SARIMA_5_ex.set_index('Date', inplace=True)
SARIMA_21_ex['Date'] = pd.to_datetime(SARIMA_21_ex['Unnamed: 0'])
SARIMA_21_ex.set_index('Date', inplace=True)

SARIMA_errors = pd.read_csv('predictions/predictions_SARIMA/SARIMA_errors.csv')
SARIMA_errors_ex = pd.read_csv('predictions/predictions_SARIMA/SARIMA_errors_ex.csv')
```

### Without Exogenous Data

```python
plt.figure(figsize=(12, 6))
plt.plot(test, label='Actual', color='black')
plt.plot(SARIMA_1['Prediction'][first_test_day:], label=f'SARIMA (H = 1)', color='dodgerblue', linestyle='dashed')   
plt.plot(SARIMA_5['Prediction'][first_test_day:], label=f'SARIMA (H = 5)', color='gold', linestyle='dashed')
plt.plot(SARIMA_21['Prediction'][first_test_day:], label=f'SARIMA (H = 21)', color='lightcoral', linestyle='dashed')
plt.legend()
plt.title('AAL Stock Price Prediction with SARIMA (without Exogenous Data)')
plt.xlabel('Year')
plt.ylabel('Price ($)')
plt.show()
```

### With Exogenous Data

```python
plt.figure(figsize=(12, 6))
plt.plot(test, label='Actual', color='black')
plt.plot(SARIMA_1_ex['Prediction'][first_test_day:], label=f'SARIMA (H = 1)', color='dodgerblue', linestyle='dashed')
plt.plot(SARIMA_5_ex['Prediction'][first_test_day:], label=f'SARIMA (H = 5)', color='gold', linestyle='dashed')
plt.plot(SARIMA_21_ex['Prediction'][first_test_day:], label=f'SARIMA (H = 21)', color='lightcoral', linestyle='dashed')
plt.legend()
plt.title('AAL Stock Price Prediction with SARIMA (with Exogenous Data)')
plt.xlabel('Year')
plt.ylabel('Price ($)')
plt.show()
```

## LSTM Predictions

```python
LSTM_1 = pd.read_csv('predictions/predictions_LSTM/H_1_window_20_usecols_[0].csv')
LSTM_5 = pd.read_csv('predictions/predictions_LSTM/H_5_window_20_usecols_[0].csv')
LSTM_21 = pd.read_csv('predictions/predictions_LSTM/H_21_window_20_usecols_[0].csv')

LSTM_1['Date'] = pd.to_datetime(LSTM_1['Date'])
LSTM_1.set_index('Date', inplace=True)
LSTM_5['Date'] = pd.to_datetime(LSTM_5['Date'])
LSTM_5.set_index('Date', inplace=True)
LSTM_21['Date'] = pd.to_datetime(LSTM_21['Date'])
LSTM_21.set_index('Date', inplace=True)

LSTM_1_ex = pd.read_csv('predictions/predictions_LSTM/H_1_window_20_usecols_[0, 2, 3, 5].csv')
LSTM_5_ex = pd.read_csv('predictions/predictions_LSTM/H_5_window_20_usecols_[0, 2, 3, 5].csv')
LSTM_21_ex = pd.read_csv('predictions/predictions_LSTM/H_21_window_20_usecols_[0, 2, 3, 5].csv')

LSTM_1_ex['Date'] = pd.to_datetime(LSTM_1_ex['Date'])
LSTM_1_ex.set_index('Date', inplace=True)
LSTM_5_ex['Date'] = pd.to_datetime(LSTM_5_ex['Date'])
LSTM_5_ex.set_index('Date', inplace=True)
LSTM_21_ex['Date'] = pd.to_datetime(LSTM_21_ex['Date'])
LSTM_21_ex.set_index('Date', inplace=True)
```

### Without Exogenous Data

```python
plt.figure(figsize=(12, 6))
plt.plot(test, label='Actual', color='black')
plt.plot(LSTM_1['Close'], label=f'LSTM (H = 1)', color='dodgerblue', linestyle='dashed')
plt.plot(LSTM_5['Close'], label=f'LSTM (H = 5)', color='gold', linestyle='dashed')
plt.plot(LSTM_21['Close'], label=f'LSTM (H = 21)', color='lightcoral', linestyle='dashed')
plt.legend()
plt.title('AAL Stock Price Prediction with LSTM (without Exogenous Data)')
plt.xlabel('Year')
plt.ylabel('Price ($)')
plt.show()
```

### With Exogenous Data

```python
plt.figure(figsize=(12, 6))
plt.plot(test, label='Actual', color='black')
plt.plot(LSTM_1_ex['Close'], label=f'LSTM (H = 1)', color='dodgerblue', linestyle='dashed')
plt.plot(LSTM_5_ex['Close'], label=f'LSTM (H = 5)', color='gold', linestyle='dashed')
plt.plot(LSTM_21_ex['Close'], label=f'LSTM (H = 21)', color='lightcoral', linestyle='dashed')
plt.legend()
plt.title('AAL Stock Price Prediction with LSTM (with Exogenous Data)')
plt.xlabel('Year')
plt.ylabel('Price ($)')
plt.show()
```

## Prophet Predictions

```python
prophet_1 = pd.read_csv('predictions/predictions_Prophet/prophet_predictions_h1.csv')
prophet_5 = pd.read_csv('predictions/predictions_Prophet/prophet_predictions_h5.csv')
prophet_21 = pd.read_csv('predictions/predictions_Prophet/prophet_predictions_h21.csv')

prophet_1['Date'] = pd.to_datetime(prophet_1['Date'])
prophet_1.set_index('Date', inplace=True)
prophet_5['Date'] = pd.to_datetime(prophet_5['Date'])
prophet_5.set_index('Date', inplace=True)
prophet_21['Date'] = pd.to_datetime(prophet_21['Date'])
prophet_21.set_index('Date', inplace=True)

prophet_1_ex = pd.read_csv('predictions/predictions_Prophet/prophet_predictions_h1_exo.csv')
prophet_5_ex = pd.read_csv('predictions/predictions_Prophet/prophet_predictions_h5_exo.csv')
prophet_21_ex = pd.read_csv('predictions/predictions_Prophet/prophet_predictions_h21_exo.csv')

prophet_1_ex['Date'] = pd.to_datetime(prophet_1_ex['Date'])
prophet_1_ex.set_index('Date', inplace=True)
prophet_5_ex['Date'] = pd.to_datetime(prophet_5_ex['Date'])
prophet_5_ex.set_index('Date', inplace=True)
prophet_21_ex['Date'] = pd.to_datetime(prophet_21_ex['Date'])
prophet_21_ex.set_index('Date', inplace=True)
```

### Without Exogenous Data

```python
plt.figure(figsize=(12, 6))
plt.plot(test, label='Actual', color='black')
plt.plot(prophet_1['Predictions'], label=f'Prophet (H = 1)', color='dodgerblue', linestyle='dashed')
plt.plot(prophet_5['Predictions'], label=f'Prophet (H = 5)', color='gold', linestyle='dashed')
plt.plot(prophet_21['Predictions'], label=f'Prophet (H = 21)', color='lightcoral', linestyle='dashed')
plt.legend()
plt.title('AAL Stock Price Prediction with Prophet (without Exogenous Data)')
plt.xlabel('Year')
plt.ylabel('Price ($)')
plt.show()
```

### With Exogenous Data

```python
plt.figure(figsize=(12, 6))
plt.plot(test, label='Actual', color='black')
plt.plot(prophet_1_ex['Predictions'], label=f'Prophet (H = 1)', color='dodgerblue', linestyle='dashed')
plt.plot(prophet_5_ex['Predictions'], label=f'Prophet (H = 5)', color='gold', linestyle='dashed')
plt.plot(prophet_21_ex['Predictions'], label=f'Prophet (H = 21)', color='lightcoral', linestyle='dashed')
plt.legend()
plt.title('AAL Stock Price Prediction with Prophet (with Exogenous Data)')
plt.xlabel('Year')
plt.ylabel('Price ($)')
plt.show()
```

## Comparison of Different Models

```python
plt.figure(figsize=(12, 6))
plt.plot(test, label='Actual', color='black')
plt.plot(SARIMA_21['Prediction'][first_test_day:], label=f'SARIMA (H = 21)', color='crimson', linestyle='dashed')
plt.plot(LSTM_21['Close'], label=f'LSTM (H = 21)', color='deepskyblue', linestyle='dashed')
plt.plot(prophet_21['Predictions'], label=f'Prophet (H = 21)', color='greenyellow', linestyle='dashed')
plt.legend()
plt.title('AAL Stock Price Prediction with SARIMA, LSTM, and Prophet with Horizon of 21 Days (without Exogenous Data)')
plt.xlabel('Year')
plt.ylabel('Price ($)')
plt.show()
```
